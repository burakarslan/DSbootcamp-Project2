```python
#İmport all required libaries first.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.model_selection import train_test_split, cross_val_score
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LassoCV, RidgeCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('stats_big5leagues.txt',sep=";", index_col='PlayerName')
```

```python
df_tm = pd.read_csv('data/tm_marketvalues.csv',sep=",", index_col='player')
```

```python
df = df.merge(df_tm,left_index=True, right_index=True)
```

```python
df
```

```python
def MorK(x):
    if(x['Value2']=='M'):
        return(1)
    else:
        return(0.001)
    
df['NetValue'] = df.apply(MorK, axis=1)
df['NetValue'] = df['Value']  * df['NetValue']
df
```

```python
df = df[df['NetValue']<100]
```

```python
fmvalue=df['NetValue']
tmvalue=df['t_value']

plt.figure(figsize=(15,15))

predicted_bar = sns.jointplot(fmvalue, tmvalue, kind="reg")
plt.xlabel ('FM 2021 values')
plt.ylabel ('Transfermarkt Values')
predicted_bar.savefig('fm vs tm values.svg')
```

```python
plt.figure(figsize=(15,15))
df["NetValue"].plot.hist(density=True, bins = 30)
df["t_value"].plot.hist(density=True, bins = 30)
predicted_bar.savefig('fm vs tm values bar chart.svg')
```
